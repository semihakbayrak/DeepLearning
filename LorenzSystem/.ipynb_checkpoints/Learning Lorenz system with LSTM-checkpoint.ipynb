```julia
using Flux, DynamicalSystems, Plots
```

```julia
ds = Systems.lorenz(œÅ = 32.0)
tr = trajectory(ds,35);
```

```julia
#Dataset
# initialize a 3D plot with 1 empty series
plt = plot3d(1, xlim=(-25,25), ylim=(-30,40), zlim=(5,60),
                title = "Lorenz Attractor", marker = 2)

for i=1:2500
    x,y,z = tr[i]
    push!(plt, x, y, z)
end

plt
```

```julia
trainset = []
for i=1:100
    data = []
    data_indeces = rand(1:2300,100)
    for i in data_indeces
        X = tr[i:i+199]
        Y = tr[i+1:i+200]
        push!(data,(X,Y))
    end
    push!(trainset,data)
end
```

```julia
testset = []
data_indeces = rand(2500:3300,10)
for i in data_indeces
    X = tr[i:i+199]
    Y = tr[i+1:i+200]
    push!(testset,(X,Y))
end
```

```julia
model = Chain(
  Dense(3,10,relu),
  LSTM(10, 20),
  Dense(20, 3))
```

```julia
#output of the model with a test instance before training
ex = model.(testset[1][1])
# initialize a 3D plot with 1 empty series
plt = plot3d(1, xlim=(-25,25), ylim=(-30,40), zlim=(5,60),
                title = "Lorenz Attractor", marker = 2)

for i=1:200
    x,y,z = Flux.Tracker.data(ex[i])
    push!(plt, x, y, z)
end

plt
```

```julia
opt = RMSProp(0.001)
function loss(x, y)
    Flux.reset!(model) #It is like reseting h
    l = sum(Flux.mse.(model.(x), y))
    #Flux.truncate!(model)
    return l
end
ps = Flux.params(model);
```

```julia
for data in trainset
    Flux.train!(loss, ps, data, opt)
end
```

```julia
#output of the model with a test instance after training
Flux.reset!(model)
ex = model.(testset[1][1])
# initialize a 3D plot with 1 empty series
plt = plot3d(1, xlim=(-25,25), ylim=(-30,40), zlim=(5,60),
                title = "Lorenz Attractor", marker = 2)

for i=1:200
    x,y,z = Flux.Tracker.data(ex[i])
    push!(plt, x, y, z)
end

plt
```
